```python
#3.번 문제 (매월 승하차 차이 계산)
data['1월승하차차'] = data['1월승차'] - data['1월하차']
data['2월승하차차'] = data['2월승차'] - data['2월하차']
data['3월승하차차'] = data['3월승차'] - data['3월하차']
data['4월승하차차'] = data['4월승차'] - data['4월하차']
data['5월승하차차'] = data['5월승차'] - data['5월하차']
```

```python
data
```

```python
#4번문제
#1월 승차인원이 가장 많은 값
data['1월승차'].max()
```

```python
#1월 승차인원이 가장 많은 정류장을 가지는 행 추출
data['1월승차'].argmax()
```

```python
#1월 승차인원이 가장 많은 정류장을 가지는 행 추출
data.iloc[data['1월승차'].argmax(),:]
```

```python
#5번 문제
#승하차 인원 차이가 가장 큰 정류장과 해당 월(1~5월중)
max_gap = abs(data.loc[:,'1월승하차차':'5월승하차차']) #절대값
```

```python
#승하차 인원 차이가 가장 큰 정류장과 해당 월(1~5월중)
#절대값을 씌우지 않은 결과
data.loc[:,'1월승하차차':'5월승하차차'].max()
```

```python
#승하차 인원 차이가 가장 큰 정류장과 해당 월(1~5월중)
##절대값을 씌운 결과
abs(data.loc[:,'1월승하차차':'5월승하차차']).max()
```

```python
#승하차 인원 차이가 가장 큰 정류장과 해당 월(1~5월중)
max_value = (abs(data.loc[:,'1월승하차차':'5월승하차차']).max()).max()
```

```python
max_value
```

```python
data_1_5_gap = abs(data.loc[:,'1월승하차차':'5월승하차차'])
```

```python
ids, cols = np.where((abs(data.loc[:,'1월승하차차':'5월승하차차']) == max_value))
#np.where: where함수는 배열에 대한 조건문,참일때 값, 거짓일때 값 의 형태로 사용, 배열을 array로 반환
#출력은 근본적으로 인덱스(위치를 반환)
```
