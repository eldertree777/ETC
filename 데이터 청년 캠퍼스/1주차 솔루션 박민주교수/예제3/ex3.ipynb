```python
import pandas as pd
import numpy as np
```

```python
#대전 지하철 역별/시간별 승하차 인원 분석 (한글인코딩 오류시 encoding='euc-kr' 추가)
data = pd.read_csv('대전광역시도시철도공사_시간대별_승하차인원_2020년01월_05월.csv',encoding='euc-kr')
```

```python
data
```

```python
#replace 함수에 regex=True 옵션을 넘기면 정규표현을 사용해 문자열을 바꿀수도 있다
data['날짜'] = data['날짜'].replace({'-':''}, regex=True)
```

```python
data
```

```python
date_list = np.unique(data['날짜']) #중복제거하여 모든 날짜 추출
```

```python
date_list
```

```python
date_list[0]
```

```python
data.loc[data['날짜'] == date_list[0],:]
```

```python
for i in date_list:
    temp = data.loc[data['날짜'] == i,:] #해달날짜 뽑아내기
    temp.to_csv("Daejeon_subway_%s.txt" %i) #날짜별 저장
```
