## 라) 데이터 전처리-시간데이터 처리

##    “처리일자”와“처리일시분초”열을 합치고, “발급일시”와 “발급시분초” 열을 합쳐서 yyyymmddhhmmss의 형식으로 만들려고 합니다. 그런데, “처리일시분초”와 “발급시분초” 열의 데이터를 확인 해보면, 0,100, 200과 같은  6자리가 아닌 다양한 값이 함께 섞여있습니다. 이를 해결하기 위해 아래와 같이 접근해보세요.
### a) “처리일시분초”와 “발급시분초” 열을 6자리 character형식으로 각각 변환하세요. 5자리 데이터는 앞에 0이 추가되어 6자리 문자로 변형됩니다. 

```python
np.sort(data_raw_non['발급시분초'].unique()) #0,100,200 과 같은 6자리가 아닌 값이 존재, 숫자형
```

```python
data_raw_non['발급일시'] = data_raw_non['발급일시'].astype(int) #발급일자에 소수점표시를 정수로 변경
data_raw_non['발급일시'] = data_raw_non['발급일시'].astype(str) #숫자형 값을 문자형으로 변경
data_raw_non['처리일자'] = data_raw_non['처리일자'].astype(str) #숫자형 값을 문자형으로 변경
```

```python
data_raw_non['처리일시분초'] = data_raw_non['처리일시분초'].astype(str) #숫자형 값을 문자형으로 변경
data_raw_non['발급시분초'] = data_raw_non['발급시분초'].astype(str) #숫자형 값을 문자형으로 변경
```

```python
data_raw_non.info() #문자로 변경되어는지 확인
```

```python
data_raw_non['발급시분초']  = data_raw_non['발급시분초'].str.zfill(width=6)   #문자로 변경후 6자리문자로 변경 빈칸에 0으로 채워짐
data_raw_non['처리일시분초']  = data_raw_non['처리일시분초'].str.zfill(width=6) #문자로 변경후 6자리문자로 변경 빈칸에 0으로 채워짐
```

```python
np.sort(data_raw_non['발급시분초'].unique()) #제대로 변경되었는지 확인
```

### b) a)에서 6자리 문자형으로 나타나는 “처리일시분초”와 “발급시분초”열을  POSIXlt(날짜,시간)으로 변경하려고 합니다. 아래 Code를 활용하여 변경하세요. 

#### b-1)

```python
data_raw_non['발급일시'].str.cat(data_raw_non['발급시분초']) #두 문자열을 붙이기
```

```python
data_raw_non['발급시간'] = data_raw_non['발급일시'].str.cat(data_raw_non['발급시분초'])  #두 문자열을 붙인 뒤 새로운 열만들기
data_raw_non['처리시간'] = data_raw_non['처리일자'].str.cat(data_raw_non['처리일시분초'])  #두 문자열을 붙이기  뒤 새로운 열만들기
```

```python
data_raw_non
```

```python
data_raw_non['처리시간'] = pd.to_datetime(data_raw_non['처리시간']) #시간형식으로 변경
data_raw_non['발급시간'] = pd.to_datetime(data_raw_non['발급시간']) #시간형식으로 변경
```
